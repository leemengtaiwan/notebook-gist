```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

num_subplots = 4
plt.figure(figsize=(10, 10))
subplot_x, subplot_y = math.ceil(num_subplots / 2), 2 # 2 means two columns

for plot in range(num_subplots):
    plt.subplot(subplot_x, subplot_y, plot+1)
    plt.scatter([x for x in range(100)], [x**plot for x in range(100)])
```
